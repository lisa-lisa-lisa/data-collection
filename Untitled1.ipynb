```python
images = {}
encodings = {}
for filename in images_dir.glob("*"):
    image = face_recognition.load_image_file(filename)
    images[str(filename.name)] = image
    try:
        encodings[str(filename.name)] = face_recognition.face_encodings(image)[0]
    except IndexError:
        pass
list(encodings.keys())   
```

```python
# plt.imshowimages["fonsi1.jpg"]
name = "bieber.jpg"

distances = face_recognition.face_distance(list(encodings.values()),encodings[name])

for i, face in enumerate(encodings):
    if distances[i] < 0.65:
        print(name, "matches",face)
distances
```
